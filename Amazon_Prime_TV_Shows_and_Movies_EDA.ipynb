## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Data Wrangling Code

# 1. Remove exact duplicate rows
titles_clean = titles.drop_duplicates().copy()
credits_clean = credits.drop_duplicates().copy()

print("Titles rows after removing duplicates:", titles_clean.shape[0])
print("Credits rows after removing duplicates:", credits_clean.shape[0])

# 2. Handle missing values for key numeric fields

# Runtime: replace missing with median runtime
titles_clean["runtime"] = titles_clean["runtime"].fillna(titles_clean["runtime"].median())

# Ratings: drop rows where BOTH imdb_score and tmdb_score are missing
titles_clean = titles_clean.dropna(subset=["imdb_score", "tmdb_score"], how="all")

# 3. Create a decade feature from release_year
titles_clean["decade"] = (titles_clean["release_year"] // 10) * 10

# 4. Parse list-like columns: genres and production_countries
import ast
import pandas as pd

def safe_parse_list(x):
    if pd.isna(x):
        return []
    if isinstance(x, list):
        return x
    try:
        value = ast.literal_eval(x)
        if isinstance(value, list):
            return value
        else:
            return [str(value)]
    except (ValueError, SyntaxError):
        return [item.strip() for item in str(x).split(",")]

titles_clean["genres_list"] = titles_clean["genres"].apply(safe_parse_list)
titles_clean["countries_list"] = titles_clean["production_countries"].apply(safe_parse_list)

# 5. Merge titles and credits for later actor/director analysis
titles_credits = titles_clean.merge(credits_clean, on="id", how="left")

titles_clean.head()
````

### What all manipulations have you done and insights you found?

Removed exact duplicate rows from both titles and credits so that each title and credit record is counted only once in the analysis.

Handled missing values by filling missing runtime with the median runtime and dropping rows where both IMDb score and TMDB score were missing, keeping the key rating fields reliable.

Created a new decade feature from release_year to study how Amazon Prime’s catalog has grown and changed across different decades.

Converted the genres and production_countries columns into clean Python lists (genres_list, countries_list) so that individual genres and countries can be exploded, counted, and visualized correctly.

Merged the cleaned titles and credits datasets on the common id field to enable analysis of actors and directors who appear most frequently on the platform.Answer Here.
